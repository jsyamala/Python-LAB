```python
#Ask the user to enter a natural number
#Print all the factors of that number

N = int(input("Enter value of N: "))
i = 1
C = 0 #this will count every factor that I find

print("The factors of",N,"are:")
while i <= N:
  #if i is a factor of N, then print i
  if N%i == 0:
    C += 1 #found +1 factor
    print(i,end=",")

  #at the end of each iteration, increment i
  i += 1

#outside while loop
print()
print(N,'has',C,"factors")
```
