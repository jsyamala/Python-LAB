```python
#creating a Flight lay out
F =[]
for R in  range(10):
  temp =[]
  for C in range(8):
    temp.append("--")
    F.append(temp)
for el in F:
  print(el)


```

```python

for el in F:
  print(el)
COUNT = 0
for row in range(len(F)):
  for col in range(len(F[0])):
    if F[row][col] == "--":
      COUNT +=1

print()      
print(COUNT,"seats available")
```

```python
#accolating seats to user

N = input("Enter first 3 letters of your name: ")
ROW = int(input("Enter chosen row index: "))
COL = int(input("Enter chosen col index: "))

if F[ROW][COL] != "--":
  print("That seat has already been booked!!""")
else:
  F[ROW][COL] = N
  print("your seat has been allocated!")
```

```python
#CANCELLATION of SEAT

N = input("Enter first 3 letters of your name: ")
ROW = int(input("Enter your seat row index: "))
COL = int(input("Enter your seat col index: "))

F[ROW][COL] = "--"
print("your seat has freed! You won't get your money back, though")
```
