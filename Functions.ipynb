```python
def my_ceiling(x):
  if int(x) ==x :
    return(x)
  elif x < 0:
    return(int(x))
  elif x>0:
    return(int(x)+1)

print(my_ceiling(-1.6))
```

```python
def my_round(x):
  str_x = str(x)
  int_x = int(x)
  if int(str_x[-1]) > 5:
    return(int_x+1)
  elif int(str_x[-1]) <= 5:
    return(int_x)
print(my_round(0.3))
```
