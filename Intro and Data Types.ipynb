```python
#program where you can modify the value of variable N
#the program will print the sum of digits of the number N

SUM = 0 #initializing SUM to 0

#You may change the value of N below
N = 100000
D = 10

while N > 0:
  Q = N//D
  R = N%D
  SUM = SUM + R
  N = Q

print(SUM)
```
