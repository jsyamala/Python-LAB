```python
def date_format(date):
    '''Input: date takes the string as input
       Output:return resultant date formats'''

    # YOUR CODE GOES HERE
    d=date.split("/")
    k=d[1]
    d[1]=d[0]
    d[0]=k
    M="/".join(d)
    
    Y=date.split("/")
    t=Y[2]
    Y[2]=Y[0]
    Y[0]=t
    S="/".join(Y)
    print(M)
    print(S)

date="08/12/1998"

print(date)
date_format(date)
```
